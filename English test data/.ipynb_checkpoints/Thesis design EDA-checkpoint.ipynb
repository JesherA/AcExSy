```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
data = pd.read_json('train.json')
data.set_index('ID',inplace=True)
```

```python
# total number of acronyms 
start = 0
for i in data['acronyms']:
    start += len(i)
    
print(start)
```

```python
# total number of Long-forms
start = 0
for i in data['long-forms']:
    start += len(i)
    
print(start)
```

```python
acronym_loc = []
for i in data['acronyms']:
    for j in i:
        acronym_loc.append(j)
        
acronym_length = pd.DataFrame(np.diff(acronym_loc), columns=['acronym_length'])

fig = px.box(acronym_length, y="acronym_length",
             width=300, height=500,
             title="Character Length Long-Form "
            )

fig.update_yaxes(dtick=2, range=[0, 19])
fig.update_layout(title='Acronym Character Length')
    # Adding Axis labels
fig.update_yaxes(title_text="<b>Acronym</b> Character Length",)

fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='#8D8A89')
fig.update_xaxes(showline=True, linewidth=2, linecolor='black')

    # Addjusting background color
fig.update_layout(paper_bgcolor='white', plot_bgcolor='white')
fig.update_traces(marker_color='#005B96')


fig.show()
```

```python
long_forms_loc = []
for i in data['long-forms']:
    for j in i:
        long_forms_loc.append(j)
        
long_forms_loc = pd.DataFrame(np.diff(long_forms_loc), columns=['long_form_length'])


fig = px.box(long_forms_loc, y="long_form_length",
             width=300, height=500,
             title="Character Length Long-Form "
            )

fig.update_yaxes(dtick=10, range=[0, 85])
fig.update_layout(title='Long-Form Character Length')
    # Adding Axis labels
fig.update_yaxes(title_text="<b>Long-Form</b> Character Length",)

fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='#8D8A89')
fig.update_xaxes(showline=True, linewidth=2, linecolor='black')

    # Addjusting background color
fig.update_layout(paper_bgcolor='white', plot_bgcolor='white')
fig.update_traces(marker_color='#38761d')


fig.show()
```
